```python
import pandas as pd # import and manage datasets
import numpy as np # mathematic library
```

```python
df = pd.read_excel('data_hackathon.xlsx')
```

```python
df.head(5)
```

```python
# create dataframe for personnel working
df_personnel = df[['report_week','hospital_code','er_staff_residents_and_rural_day_on_call','er_staff_specialist_day_on_call','er_staff_mic_day_on_call','er_staff_nurse_day_on_call','er_staff_non_professional_nurse_day_on_call','er_staff_residents_and_rural_night_on_call','er_staff_specialist_night_on_call','er_staff_mic_night_on_call','er_staff_nurse_night_on_call','er_staff_non_professional_nurse_night_on_call']].copy()
```

```python
df_personnel.head(5)
```

```python
# check if the columns have null values --- personnel on day shift
print(df_personnel['er_staff_residents_and_rural_day_on_call'].isnull().sum())
print(df_personnel['er_staff_specialist_day_on_call'].isnull().sum())
print(df_personnel['er_staff_mic_day_on_call'].isnull().sum())
print(df_personnel['er_staff_nurse_day_on_call'].isnull().sum())
print(df_personnel['er_staff_non_professional_nurse_day_on_call'].isnull().sum())
```

```python
# check if the columns have null values --- personnel on night shift
print(df_personnel['er_staff_residents_and_rural_night_on_call'].isnull().sum())
print(df_personnel['er_staff_specialist_night_on_call'].isnull().sum())
print(df_personnel['er_staff_mic_night_on_call'].isnull().sum())
print(df_personnel['er_staff_nurse_night_on_call'].isnull().sum())
print(df_personnel['er_staff_non_professional_nurse_night_on_call'].isnull().sum())
```

```python
er_day_on_call = []
er_night_on_call = []
for i in range(len(df_personnel)):
    x = df_personnel['er_staff_residents_and_rural_day_on_call'][i]+ df_personnel['er_staff_specialist_day_on_call'][i]+ df_personnel['er_staff_mic_day_on_call'][i]+ df_personnel['er_staff_nurse_day_on_call'][i]+ df_personnel['er_staff_non_professional_nurse_day_on_call'][i] 
    er_day_on_call.append(x)
    y = df_personnel['er_staff_residents_and_rural_night_on_call'][i]+ df_personnel['er_staff_specialist_night_on_call'][i]+ df_personnel['er_staff_mic_night_on_call'][i]+ df_personnel['er_staff_nurse_night_on_call'][i]+ df_personnel['er_staff_non_professional_nurse_night_on_call'][i] 
    er_night_on_call.append(y)
```

```python
df_personnel['er_day_on_call'] = er_day_on_call
df_personnel['er_night_on_call'] = er_night_on_call
```
